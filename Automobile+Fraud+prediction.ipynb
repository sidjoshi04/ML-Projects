```python
# Replace '?' with '0'
df['property_damage'].replace(to_replace='YES', value=1, inplace=True)
df['property_damage'].replace(to_replace='NO', value=0, inplace=True)
df['property_damage'].replace(to_replace='?', value=0, inplace=True)
df['police_report_available'].replace(to_replace='YES', value=1, inplace=True)
df['police_report_available'].replace(to_replace='NO', value=0, inplace=True)
df['police_report_available'].replace(to_replace='?', value=0, inplace=True)
```

```python
#Converting  multi-class type variable collision_type using LabelEncoder,
#'?' will be converted to '0'
le = LabelEncoder()
df['collision_type']=le.fit_transform(df['collision_type'])
```

```python
dummies= pd.get_dummies(df[[
    'policy_state', 'policy_csl', 'insured_sex', 'insured_education_level',
       'insured_occupation', 'insured_hobbies', 'insured_relationship',
       'incident_type', 'incident_severity', 'authorities_contacted',
       'incident_state', 'incident_city', 'auto_make', 'auto_model']])
```

```python
dummies.head(5)
```

```python
dummies.shape
```

```python
X = pd.concat([dummies, df._get_numeric_data()], axis=1)  # joining numeric columns
X.head(5)
```

```python
X.shape
```

```python
y = X['fraud_reported']
```

```python
x = X.drop(['fraud_reported'],axis=1)
```

```python
x.head()
```

```python
x.shape,y.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(x)
```

```python
x = pd.DataFrame(X, columns=x.columns)
```

```python
x.head(5)
```

```python
x.plot(kind='hist')
plt.show()
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 101)
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
prediction = lr.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, prediction))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(x_train,y_train)
knn_pred=knn.predict(x_test)
```

```python
knn_pred=knn.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, knn_pred))
```

```python
#Desicion Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
features = list(x_train.columns)
fi = pd.DataFrame({'feature': features,
                   'importance': dt.feature_importances_}).\
                    sort_values('importance', ascending = False)
fi.head(10)
```
